### 3-dimensional plots of pixel features

#### Lettuce plain weak - RGB

```python
plt.rcParams['figure.figsize'] = [10, 10]

in_img = lettuce_plainweak

r, g, b = cv.split(in_img)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
pixel_colors = in_img.reshape((np.shape(in_img)[0]*np.shape(in_img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

### Image masking with ranges

```python
light_white = (40, 50, 0)
dark_white = (100, 250, 250)

mask_white = cv.inRange(lettuce_plainweak, light_white, dark_white)

final_mask = mask_white

final_result = cv.bitwise_and(lettuce_plainweak, lettuce_plainweak, mask=final_mask)
plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.show()
```

```python
blur = cv.GaussianBlur(final_result, (7, 7), 0)
plt.imshow(blur)
plt.show()
```
